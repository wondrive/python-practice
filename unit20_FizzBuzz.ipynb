문제1. 3의 배수일 때와 5의 배수일때 출력 (1-100 사이 숫자)

```python
for i in range(1, 101):
    if i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

문제2. 3과 5의 공배수 처리하기

```python
for i in range(1, 101):
    if i%3==0 and i%5==0:
        print('FizzBuzz')
    elif i%3==0:
        print('Fizz')
    elif i%5==0:
        print('Buzz')
    else:
        print(i)
```

같은 코드 단축하기 => 너무 신기하다!

```python
for i in range(1, 101):
    print('Fizz' * (i%3==0) + 'Buzz' * (i%5==0) or i)
```

# 연습문제

문제1. 2, 11의 배수, 공배수 처리

```python
# 1-100 사이 숫자
for i in range(1, 101):
    if i%2==0 and i%11==0:
        print('FizzBuzz')
    elif i%2==0:
        print('Fizz')
    elif i%11==0:
        print('Buzz')
    else:
        print(i)
```

문제2. 5와 7의 배수, 공배수 처리

```python
start, stop = list(map(int, input().split()))

for i in range(start, stop+1):
    print('Fizz' * (i%5==0) + 'Buzz' * (i%7==0) or i)
```
