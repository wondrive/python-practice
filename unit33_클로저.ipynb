# 1. 지역변수, 전역변수

```python
x = 10      # 전역변수

def foo():
    print(x, '1')

foo()
print(x, '2')
```

```python
def foo():
  y = 10    # 지역변수
  print(y)

foo()
print(y)
```

- 함수 안에서 전역변수 변경

```python
y = 10
def foo():
  global y      # 지역변수 -> 전역변수 변경
  y = 20
  print(y)

foo()
print(y)
```

```python
def foo():
  global y      # 전역변수 없는 상태에서 생성
  y = 20
  print(y)

foo()
print(y)
```

- 네임스페이스(name space, 이름 공간) 확인:

# 2. 함수 안에서 함수 만들기

```python
def A():
  x = 10
  def B():
    x = 20

  B()
  print(x)

A()
```

B의 x는 또다른 지역변수일뿐.

- nonlocal 지역변수
  - 지역변수 찾을때 가장 가까운 함수(지역)부터 찾음

```python
# 예제 1
def A():
  x = 10
  def B():
    nonlocal x
    x = 20

  B()
  print(x)

A()
```

```python
# 예제 2
def A():
  x = 10
  y = 100

  def B():
    x = 20

    def C():
      nonlocal x
      nonlocal y
      x += 30
      y += 300
      print(x)
      print(y)
    C()

  B()

A()
```

- 전역변수 사용하기

```python
x = 1

def A():
  x = 10
  def B():
    x = 20
    def C():
      global x
      x += 30
      print(x)
    C()
  B()
A()
```

전역변수는 가급적 사용하지 않는 게 좋음. 코드가 복잡해지면 어느 함수에서 바꾸는지 모르기 때문임

# 3. 클로저 사용하기

- 클로저란 지역변수, 코드 등을 유지하다가 함수를 호출할 때 다시 꺼내서 사용하는 함수
- 다음의 경우에 사용
  - 지역변수와 코드를 묶어서 사용하고 싶을 때 활용
  - 데이터를 숨기고 싶을때 사용: 클로저에 속한 지역 변수는 바깥에서 접근할 수 없음

```python
def calc():
  a = 3
  b = 5
  def mul_add(x):
    return a * x + b
  return mul_add

c = calc()
print(c(1), c(2), c(3), c(4), c(5))
```

- lambda로 클로저 만들기

```python
def calc():
  a = 3
  b = 5
  return lambda x: a * x + b

c = calc()
print(c(1), c(2), c(3), c(4), c(5))
```

- 클로저의 지역변수 변경하기  
  - nonlocal: 클로저의 지역변수를 변경하고 싶을 때 사용함

```python
def calc():
  a = 3
  b = 5
  total = 0

  def mul_add(x):
    nonlocal total    # nonlocal: 바깥에 있는 지역변수 사용. 클로저의 지역변수를 변경하고 싶을 때 사용함
    total += a * x + b
    print(total)
  return mul_add

c = calc()
c(1)
c(2)
c(3)
```

# 연습문제

```python
# 다음 소스 완성. 함수 c를 호출할 때마다 호출 횟수가 출력되게 만들기
def counter():
  i = 0
  def count():
    nonlocal i
    i += 1
    return i
  return count

c = counter()
for i in range(10):
  print(c(), end=' ')
```

# 심사문제  
다음 소스를 완성하여 호출할 때마다 숫자가 1씩 줄어들게 만드세요. 클로저 함수 사용할것.

```python
def countdown(n):
  total = n + 1
  def count():
    nonlocal total
    total -= 1
    return total
  return count

n = int(input())

c = countdown(n)
for i in range(n):
  print(c(), end=' ')
```
