# 1. 람다 표현식으로 함수 만들기
- **lambda expression** 또는 **anonymous function** 이라고도 함

```python
# 기존 함수
def plus_ten(x):
  return x+10

plus_ten(1)
```

```python
# 람다 함수1.
plus_ten = lambda x: x+10
plus_ten(1)

# 람다 함수2.
(lambda x: x+10)(1)
```

```python
# 매개변수가 없는 람다함수
(lambda: 1)()   # return값이 1인 람다함수
```

- 람다 함수 내에서는 변수 만들 수 없음

```python
# 단, 람다 함수 밖의 함수는 사용 가능
y = 10
(lambda x: x+y)(1)
```

- 람다 함수를 인수로 사용하기

```python
# 일반 함수
def plus_ten(x):
  return x+10

# 람다 함수
plus_ten2 = lambda x: x+10

print(list(map(plus_ten, [1, 2, 3])))
print(list(map(plus_ten2, [1, 2, 3])))
```

# 2. map, filter, recude 함수와 함께 활용하기

1. if: 조건부 표현식 사용하기
- 반드시 else 사용
- elif 사용 불가 => if-else 문을 여러번 사용하는 걸로 대체 가능 => 이런 경우는 그냥 def로 함수 만드는 게 가독성 좋음

```python
a = list(range(1, 11))
list(map(lambda x: str(x) if x%3==0 else x, a))    # 3의 배수를 문자열로 변환
```

```python
list(map(lambda x: str(x) if x%3==0 else float(x) if x%5==0 else x, a))    # 3의 배수를 문자열로 변환 + 5의 배수는 실수형
```

2. map에 여러 객체 넣기  
- map에는 iterable 객체를 여러 개 넣을 수 있음

```python
a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]

list(map(lambda x, y: x*y, a, b))
```

3. filter 사용
- 반환값이 True인 요소만 가져옴

```python
def f(x):
  return x>5 and x<10

a = [8, 3, 2, 10, 15, 7, 1, 9, 0, 11]
list(filter(f, a))    # 거짓인 요소는 버리고 출력
```

4. reduce
- reduce(함수, iterable 객체)  
- 각 요소를 함수로 처리한 후 이전 결과와 누적해서 반환하는 함수

```python
from functools import reduce

def f(x, y):
  return x+y

a = [1, 2, 3, 4, 5]
reduce(f, a)        # 처음엔 (1, 2)가 인수로 들어가고, 그 다음엔 (1+2, 3), -> (1+2+3, 4) -> ...
```

```python
# 위 코드를 람다함수로 작성
reduce(lambda x, y: x+y, a)
```

# 연습문제
다음 소스를 완성하시오. 확장자가 .jpg, .png인 파일만 리스트로 출력되게 만드세요. 확장자 검사시 문자열 메소들 활용하세요.

```python
# 내 정답
files = ['font', '1.png', '10.jpg', '11.gif', '2.jpg', '3.png', 'table.xslx', 'spec.docx']

print(list(filter(lambda x: '.jpg' in x or '.png' in x, files)))
```

```python
# 책 정답
print(list(filter(lambda x: x.find('.jpg')!=-1 or x.find('.png')!=-1, files)))
```
