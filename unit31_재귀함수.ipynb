- Recursive call (재귀호출): 함수 안에서 자기 자신을 호출하는 것

# 1. 재귀호출 사용하기

```python
def hello():
    print('hello world')
    hello()

hello()
```

ㄴ 재귀호출이 종료되지 않아 최대 재귀 깊이를 초과하여 에러(Recursion Error) 발생

# 2. 재귀호출로 팩토리얼 구하기

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

factorial(5)
```

# 연습문제

어렵다.. 해설지 보고 풀었다.

```python
# 다음 소스를 완성하여 문자열이 회문인지 판별하고, 결과를 True/False로 출력되게 만드세요. (단, 재귀호출 사용할것)
def is_palindrome(word):
    # 1. 종료조건
    if len(word) < 2:
        return True
    if word[0] != word[-1]:
        return False
    
    return is_palindrome(word[1:-1])    # return 써줘야 하는 이유: 안 써주면 맨 마지막 재귀호출 함수의 True를 맨 첫 재귀호출 함수의 return 값으로 가져올 수 없음

print(is_palindrome('hello'))
print(is_palindrome('level'))
```

# 심사문제

문제: 재귀호출로 피보나치 수 구하기. 정수 한 개를 입력(범위: 10~30)하여 다음 소스코드 완성. 피보나치 수는 0과 1로 시작함.

```python
def fibonacci(count, num1=0, num2=1):
    # 종료 조건:
    if count == 1:  # 덧셈을 n번 하는 거니까 n번째 (거꾸로 세면 1번째) 때 최종 덧셈값 반환
        return f'{num2}'
    return fibonacci(count=count-1, num1=num2, num2=num1+num2) # 다음 수 넘겨줘야 함

n = int(input())
print(fibonacci(n))
```
